## 4. Update "Abel's" CGPA to 4.0

```python
#BMI Calculator

Weight = float(input("Enter your Weight"))
Height = float(input("Enter your Height"))

BMI = Weight/(Height**2)
print("Your BMI is:" ,BMI)


if BMI < 18.5:
    print("Underweight")
elif BMI >18 and BMI<25:
    print("Normal")
else:
    print("Obse")
```

```python
x = 20


# Tree leaves

for y in range(x):
    spaces = " " * (x -y -1)
    stars = "*" * (2 * y + 1)
    print(spaces + stars)

#Tree Truck

for y in range(x):
    spaces =  " " *(x-1)
    print(spaces + "*")
```

```python
# Arbitrary Arguments
# *args: collect extra positional (non-keyword) arguments as a tuple.
# **Kwargs: collects extra keyword arguments as a dictionary.

def myFun(*args, **Kwargs):
    print("Non-keyword Arguments (*args:)")
    for arg in args:
        print(arg)
    print("\nKeyword Arguments(**Kwargs):")
    for key, value in Kwargs.items():
        print(f"{key} == {value}")
myFun('Hey', 'Welcome', first='Python', mid='for', last='python')
```

```python
# Nested Functions / Inner function (A function (f1) defined and called another function (f2))

def f1():
    s = 'I love python'
    def f2():
        print("You need to call me")
        print(s)
    f2()
f1()
    
```

```python
f1()
```

```python
# Anonymous Funtion

# def keyword is used to define the normal functions 
# lambda keyword is used to create anonymous functions

def c1(x):return x*x*x
c2 = lambda x:x*x*x
print(c1(7))
print(c2(7))
```

```python
# Write a function which accepts Basic Salary of an employee and prints (tax,pension, netpay)
# A function that takes a single argument called salary and the function arguments are (tax,pension, netpay)
# Use Ethiopian Tax System net_pay = Basic Salary - (tax + pension)
```

```python
def sales(tax =0.15):
    
    basic_salary =float( input("Enter Basic Salary:"))
    print("Basic Salary is:", basic_salary)
    pension = float(input("Enter pension:"))
    print("Pension is:", pension)
    
    calculated_tax = basic_salary * tax
    net_pay = basic_salary-(calculated_tax + pension)
    return net_pay

    
print("Net pay is:", sales())

```
