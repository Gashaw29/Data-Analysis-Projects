## GHC QR Code

```python


import pandas as pd
import qrcode 
from PIL import Image, ImageDraw, ImageFont
import os

# READ EXCEL FILE

data = pd.read_excel("2018 GHC HPs Info.xlsx")

print("Excel file loaded successfully\n")


# Clean column names

data.columns = data.columns.str.strip()
output_folder = "GHC_QR_Codes"


if not os.path.exists(output_folder):
    os.makedirs(output_folder)

    


for index, row in data.iterrows():
 
    name = row["Fullname_Amharic"]
    eng_name = row["Fullname_English"]
    sex = row["Sex"]
    department = row["Department"]
    level = row["Level"]
    code = row["Code"]
    phone = row["Phone"]
    photo = row["Photo"]
    
    qr_text = (
    f"Fullname_Amharic:{name}\n"
    f"Fullname_English:{eng_name}\n"
    f"Sex:{sex}\n"
    f"Department:{department}\n"
    f"Level:{level}\n"
    f"Code:{code}\n"
    f"Phone:{phone}\n"
    f"Photo:{photo}\n"
    )
    
    



    qr_image = qrcode.make(qr_text)

    filename = f"{eng_name}.png"


    filepath = os.path.join(output_folder,filename)

    qr_image.save(filepath)

    print(f"{filename} generated successfully")

    print ("\n All QR codes generated!")
```

```python
import pandas as pd
import qrcode 
from PIL import Image, ImageDraw, ImageFont
import os

# READ EXCEL FILE

data = pd.read_excel("2018 GHC HPs Info.xlsx")

print("Excel file loaded successfully\n")


# Clean column names

data.columns = data.columns.str.strip()
output_folder = "GHC_ID_CARD"


if not os.path.exists(output_folder):
    os.makedirs(output_folder)

    


for index, row in data.iterrows():
 
   
    eng_name = row["Fullname_English"]
    department = row["Department"]
    level = row["Level"]
    code = row["Code"]
    phone = row["Phone"]
    date = row["H_Date"]
    health_center = row["Health Center"]
    photo = row["Photo"]



      # CREATE QR CODE
    # ----------------------------------------
    
    qr_data = f"Fullname_English:{eng_name}"
    qr_data = f"\nHealth Center:{health_center}"
    
    qr = qrcode.make(qr_data)


 # Resize QR
    qr = qr.resize((150, 150))



    
 # CREATE ID CARD
    # ----------------------------------------
    
    card = Image.new("RGB", (600, 350),
                     color="white")
    
    draw = ImageDraw.Draw(card)


    font = ImageFont.load_default()



 # LOAD PHOTO
    # ----------------------------------------
    
    photo_path = os.path.join("photos",
                              photo)
    
    photo = Image.open(photo_path)
    
    photo = photo.resize((150, 180))
    
    # Paste photo
    card.paste(photo, (30, 80))


    
    # PASTE QR CODE
    # ----------------------------------------
    
    card.paste(qr, (420, 170))





    # Default font
    font = ImageFont.load_default()


     # FIXED TEXT LAYOUT (IMPORTANT)
    # -----------------------------

    x_text = 200
    y_start = 40
    line_gap = 40   # spacing between lines

    
    
    

    draw.text((x_text, y_start + line_gap),
              f"Fullname_English: {eng_name}",
              fill="black",
              font=font)
    
    draw.text((x_text, y_start + line_gap),
              f"Department: {department}",
              fill="black",
              font=font)
    
    draw.text((x_text, y_start + 2 * line_gap),
              f"Level: {level}",
              fill="black",
              font=font)
    
    
    draw.text((x_text, y_start + 2 * line_gap),
              f"Code: {code}",
              fill="black",
              font=font)
    draw.text((x_text, y_start + 2 * line_gap),
              f"Hire Date: {date}",
              fill="black",
              font=font)



  # SAVE CARD

    filename = f"{eng_name}.png"
    
    filepath = os.path.join(output_folder,
                            filename)
    
    card.save(filepath)
    
    print(f"{filename} created")

    print("\nAll ID cards generated!")
    
    
```
