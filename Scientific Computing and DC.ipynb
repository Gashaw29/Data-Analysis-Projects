## Data Cleaning Techniques (Removal, Imputation & Flagging)

```python
# Create a dataset

info = {'ID':['Eth - 101', 'Eth - 102', 'Eth - 103', 'Eth - 104', 'Eth - 105'],
        'Name':['Haileslassie Zeferu Abel', 'Worku Abenezer Girma', 'Abenezer Reta Tamirat', 'Abraham Melisew Fireda', 'Hailu Abraham Niguse'],
        'Age' : [25,None,29,33,None],
        'Salary':[55000, 48000, None, 62000,None]}

df = pd.DataFrame(info)
print(df)
```

```python
# Method 1: Removal : dropna()

df_cleaned = df.dropna()
print(df_cleaned)
```

```python
# Method2: Imputation : Impute missing values w the mean of the column : fillna()

for column in df.columns:
    if df[column].dtype == 'float64' or df[column].dtype == 'int64':
        df[column]= df[column].fillna(df[column].mean())

print(df)
```

```python
# Method 3: Flagging

info = {'ID':['Eth - 101', 'Eth - 102', 'Eth - 103', 'Eth - 104', 'Eth - 105'],
        'Name':['Haileslassie Zeferu Abel', 'Worku Abenezer Girma', 'Abenezer Reta Tamirat', 'Abraham Melisew Fireda', 'Hailu Abraham Niguse'],
        'Age' : [25,None,29,33,None],
        'Salary':[55000, 48000, None, 62000,None]}

df = pd.DataFrame(info)


# Create a column-based flag
for column in df.columns:

    if df[column].isnull().any():
        df[f'{column} Flag'] = df [column].isna().apply(lambda x: 'Missing' if x else 'Not Missing')

# Create a row-based flag
df['Missing_Flag'] = df.isnull().any(axis=1).apply(lambda x: 'Missing' if x else 'Not Missing')



print(df)
```

#### Remove Duplicates

```python
# Sample dataset
data = {'ID': ['Eth - 101', 'Eth - 102', 'Eth - 103', 'Eth - 104', 'Eth - 105', 'Eth - 102', 'Eth - 104', 'Eth - 106'],
        'Name': ['Haileslassie Zeferu Abel', 'Worku Abenezer Girma', 'Abenezer Reta Tamirat', 'Abraham Melisew Fireda', 'Hailu Abraham Niguse', 
'Worku Abenezer Girma', 'Abraham Melisew Fireda', 'Hailu Abraham Niguse'],
        'Age': [25, 29, 29, 33, 29, 29, 33, 29],
        'Salary': [55000, 48000, 55000, 62000, 55000, 48000, 62000, 55000],
        'City': ['Addis Ababa', 'Adama', 'Addis Ababa', 'Adama', 'Addis Ababa', 'Adama', 'Adama', 'Addis Ababa']}
        
# Original Dataset
df = pd.DataFrame(data)
print("Original Dataset")
print(df, '\n')
```

```python
# EXact Match - based on all columns

df_cleaned = df.drop_duplicates()

print("Duplicates Removed - Exact Match")
print(df_cleaned, '\n')
```

```python
# Partial Match - based on specific columns only (e.g., 'ID')

df_cleaned_id = df.drop_duplicates(subset=['ID'])

print("Duplicates Removed - Partial Match based on ID")
print(df_cleaned_id, '\n')
```

```python
# Partial Match - based on specific columns only (e.g., 'Name')

df_cleaned_name = df.drop_duplicates(subset=['Name'])

print("Duplicates Removed - Partial Match based on Name")
print(df_cleaned_name, '\n')
```

```python
# Partial Match - based on specific columns only (e.g., 'ID' and 'Name')

df_cleaned_specific = df.drop_duplicates(subset=['ID','Name'])

print("Duplicates Removed - Partial Match based on ID and Name")
print(df_cleaned_specific, '\n')
```

#### Normalizing Data Formats

```python
from dateutil.parser import parse
```

```python
event  = {'Event ID' : ['E-001', 'E-002', 'E-003', 'E004', 'E005'],
        'Event Date' : ['01/12/2023', 'December 2, 2023', '2023-12-03', '12.04.2023', '5th Dec 2023'],
        'Product Name' : ['apple', 'Banana', 'ORANGE', 'BaNAna', 'Apple'],
        'Weight' : ['70 kg', '100 lbs', '88 KG', '180 pound', '94 kilograms']
}

df = pd.DataFrame(event)

print("Original Dataset\n", df)
```

```python
# Custom function to parese dates

def parse_date(date_str):


    try:
        return parse(date_str, dayfirst=False)

    except ValueError:
        return pd.NaT   # Pandas Not a Time


# Apply the custom function

df['Event Date'] = df['Event Date'].apply(parse_date)


# Standardize format to YYYY-MM-DD

df['Event Date'] = df['Event Date'].dt.strftime('%Y-%m-%d')

print("Result", df)
```

```python
# Text Case Standardization


df = pd.DataFrame(event)

# Convert 'Prouct Name to' to lowercase

df['Product Name'] = df['Product Name'].str.lower()

print(df)
```

```python
# Unit Conversion

df = pd.DataFrame(event)

# Extract numneric value and unit

df['Numeric Value'] = df['Weight'].str.extract(r'(\d+\.?\d*)').astype(float)

# Use str to convert the extracted unit to lowercase

df['Unit'] = df['Weight'].str.extract(r'([a-zA-Z]+)') [0].str.lower()

# Conversion function

def convert_to_kg(row):
    if row['Unit'] in ['kg', 'kilograms']:
        return row['Numeric Value']

    elif row['Unit'] in ['lbs', 'pound']:
        return row ['Numeric Value'] * 0.453592

    else:
        return None # or handle unknown units

df['Weight in kg'] = df.apply(convert_to_kg, axis=1)

print(df[['Event ID', 'Weight', 'Weight in kg']])
```

#### Outlier Detection and Handling  ... ( Visual Inspection, Statistical Mehtods, Decision on Handling)

```python
import pandas as pd 
import matplotlib.pyplot as plt

employee = {'Person ID': range(1,13),

            'Full Name': ['Haileslassie Zeferu Abel', 'Worku Abenezer Girma', 'Abenezer Reta Tamirat', 'Abraham Melisew Fireda', 'Hailu Abraham Niguse', 'Weldetinsae Adonay Gebrehiwot', 'Debebe Gebremeskel Akin', 'Teferi Amanuel Abebaw', 'Muche Amanuel Kahsu', 'Amanuel Degulaw Zerga', 'Amanuel Abate Nega', 'Amanuel Solomon Hailu'],
            'Income': [80000, 95000, 11500, 75000, 110000, 85000,100000,  199000,105000,90000,80000,95000]

           }

df = pd.DataFrame(employee)


# Box Plot


plt.figure(figsize = (8,4))

plt.boxplot(df['Income'], vert = False, showfliers = True)

plt.title('Box Plot of Incomes')
plt.xlabel('Income ($000)')

plt.show()
```

```python
# Scatter Plot 
plt.figure(figsize=(8, 4)) 
plt.scatter(df['Person ID'], df['Income']) 
plt.title('Scatter Plot of Incomes') 
plt.xlabel('Person ID') 
plt.ylabel('Income ($000)') 
plt.show() 
```

```python
df = pd.DataFrame(employee)


# Calculate Q1, Q3, bounds and identity outliers

Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)

IQR = Q3-Q1

lower_bound = Q1 - 1.5 * IQR

upper_bound = Q3 + 1.5 * IQR


df['Outlier'] = df ['Income'].apply(lambda x: 'Outlier' if x > upper_bound else 'Normal')


print(df)
```
