```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

## Web Scrapping 

```python
# Define the targer URL

URL = "https://quotes.toscrape.com/"
```

```python
# Download the web page

print("Downloading web page...")

# Send http get request
response = requests.get(URL)
```

```python
if response.status_code == 200:
    print("Success! Parsing HTML...")
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
scraped_data = []
```

```python
quote_elements = soup.find_all('div', class_='quote')
```

```python
for element in quote_elements:
    text = element.find('span', class_= 'text').text
    author = element.find('small', class_= 'author').text
    tags = [tag.text for tag in element.find_all('a', class_='tag')]

    scraped_data.append({
    'Quote': text,
    'Author': author,
    'Tags': ',' .join(tags)
    })
    df = pd.DataFrame(scraped_data)
    df.to_csv('scraped_quotes.csv', index=False)
    print("Done! Data successfully saved to 'scraped_)quotes.csv'")

else:
 print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
quotes = soup.find_all("span", class_="text")
for quote in quotes:
    print(quote.text)
```
