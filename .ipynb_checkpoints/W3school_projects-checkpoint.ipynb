## 1.Calculator

```python
# Takes user input and converts input into decimal
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))


# Operators with error handling and debugging
print("\nChoose Operation")
print("1 = Addition")
print("2 = Subtraction")
print("3 = Multiplication")
print("4 = Division")
print("5 = Modulus")
print("6 = Power")



choice = input("Enter choice (1/2/3/4/5/6): ")


# Conditionals for decision making

if choice == "1":
    result = num1 + num2
    print("\nResult =",result)

elif choice == "2":
    result = num1 - num2
    print("\nResult = " ,result)

elif choice == "3":
    result = num1 * num2
    print("\nResult = ", result)

elif choice == "4":
   if num2 != 0:
       result = num1/num2
       print("\nResult =", result)

   else:
       print("\nError: Cannot divide by zero")

elif choice == "5":
    result = num1%num2
    print("\nResult = ", result)

elif choice == "6":
    result = num1**num2
    print("\nResult = " , result)

else:
    print("\nInvalid choice")
```

## 5.Password Generator

```python
#Import random module

import random

letters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"

numbers = "0123456789"

symbols = "!@#$%^&*()?"

all_characters = letters + numbers + symbols

length = int(input("Enter password lenght:"))

password = ""


for i in range(length):
    random_character = random.choice(all_characters)

    password += random_character

print("\nGenerated Password:")
print(password)
    

```

```python
#Professional password Generator



import random
import string


def generate_password(length):

    all_characters = (
        string.ascii_letters +
        string.digits +
        string.punctuation
    )


    password = ""

    for i in range(length):
        character = random.choice(all_characters)

        password += character

        return password


        
try:

    length = int(input("Enter password length:"))

    if length < 4:
        print("\nPassword should be at least 4 characters.")

    else:

        new_password = generate_password(length)

        print("\nGenerated password:")
        print(new_password)

except:
    print("\nInvalid input: please enter numbers only.")

```

```python
import pandas as pd
import qrcode 
import os


output_folder = "GHC_QR_Codes"


if not os.path.exists(output_folder):
    os.makedirs(output_folder)

    

data = pd.read_excel("2018 GHC HPs Info.xlsx")

print("Excel file loaded successfully\n")



for index, row in data.iterrows():
 
    name = row["Fullname_Amharic"]
    eng_name = row["Fullname_English"]
    sex = row["Sex"]
    department = row["Department"]
    level = row["Level"]
    code = row["Code"]
    phone = row["Phone"]
    
    qr_text = (
    f"Fullname_Amharic:{name}\n"
    f"Fullname_English:{eng_name}\n"
    f"Sex:{sex}\n"
    f"Department:{department}\n"
    f"Level:{level}\n"
    f"Code:{code}\n"
    f"Phone:{phone}\n"
    )
    
    



    qr_image = qrcode.make(qr_text)

    filename = f"{eng_name}.png"


    filepath = os.path.join(output_folder,filename)

    qr_image.save(filepath)

    print(f"{filename} generated successfully")

    print ("\n All QR codes generated!")
```
