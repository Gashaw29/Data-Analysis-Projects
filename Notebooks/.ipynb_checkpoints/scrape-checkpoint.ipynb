```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL to scrape
url = "https://quotes.toscrape.com/"

# Send request
headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

# Check request status
if response.status_code == 200:
    print("Page loaded successfully")
else:
    print("Failed to load page:", response.status_code)

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Extract quote blocks
quotes = soup.find_all("div", class_="quote")

data = []

for quote in quotes:
    text = quote.find("span", class_="text").get_text(strip=True)
    author = quote.find("small", class_="author").get_text(strip=True)

    tags = [tag.get_text(strip=True) for tag in quote.find_all("a", class_="tag")]

    data.append({
        "quote": text,
        "author": author,
        "tags": ", ".join(tags)
    })

# Convert to DataFrame
df_quotes = pd.DataFrame(data)

print(df_quotes.head())

# Save to CSV
df_quotes.to_csv("scrapes/scraped_quotes.csv", index=False)
```

# Scraping Multiple Pages 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://quotes.toscrape.com/page/{}/"

all_data = []

headers = {
    "User-Agent": "Mozilla/5.0"
}

for page in range(1, 11):
    url = base_url.format(page)

    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print(f"Page {page} failed")
        continue

    soup = BeautifulSoup(response.text, "html.parser")

    quotes = soup.find_all("div", class_="quote")

    for quote in quotes:
        text = quote.find("span", class_="text").get_text(strip=True)
        author = quote.find("small", class_="author").get_text(strip=True)

        tags = [
            tag.get_text(strip=True)
            for tag in quote.find_all("a", class_="tag")
        ]

        all_data.append({
            "page": page,
            "quote": text,
            "author": author,
            "tags": ", ".join(tags)
        })

df = pd.DataFrame(all_data)

print(df.head())
print(df.shape)

df.to_csv("scrapes/all_quotes.csv", index=False)
```

## Scraping Addis Admas 

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
import time

BASE_URL = "https://www.addisadmassnews.com/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
}

def scrape_addis_admass_home():
    response = requests.get(BASE_URL, headers=headers, timeout=20)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    articles = []

    # Addis Admass homepage article titles are mostly links inside item/article blocks
    links = soup.find_all("a", href=True)

    for link in links:
        title = link.get_text(strip=True) # process link one by one and strip removes extra spaces
        href = link.get("href")  # Extract URL

        if not title or len(title) < 10:
            continue

        full_url = urljoin(BASE_URL, href)

        # keep only internal article links
        if "addisadmassnews.com" not in full_url:
            continue

        parent_text = link.find_parent().get_text(" ", strip=True) if link.find_parent() else ""

        created_date = None
        if "Created on" in parent_text:
            created_date = parent_text.split("Created on")[-1].strip()[:20]

        articles.append({
            "title": title,
            "url": full_url,
            "created_date": created_date,
            "snippet": parent_text[:300]
        })

    df = pd.DataFrame(articles)

    # remove duplicates
    df = df.drop_duplicates(subset=["title", "url"]).reset_index(drop=True)

    return df


df_articles = scrape_addis_admass_home()

print(df_articles.head())
print(df_articles.shape)

df_articles.to_csv("scrapes/addis_admass_articles.csv", index=False, encoding="utf-8-sig")
```

## Multiple Category Pages 

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
import time

BASE_URL = "https://www.addisadmassnews.com/"

headers = {
    "User-Agent": "Mozilla/5.0"
}

def get_soup(url):
    try:
        response = requests.get(url, headers=headers, timeout=20)

        if response.status_code != 200:
            print(f"Skipping {url} | status code: {response.status_code}")
            return None

        return BeautifulSoup(response.text, "html.parser")

    except requests.exceptions.RequestException as e:
        print(f"Error loading {url}: {e}")
        return None


def scrape_links_from_page(url):
    soup = get_soup(url)

    if soup is None:
        return []

    data = []

    for link in soup.find_all("a", href=True):
        title = link.get_text(strip=True)
        href = link.get("href")

        if not title or len(title) < 8:
            continue

        full_url = urljoin(BASE_URL, href)

        if "addisadmassnews.com" not in full_url:
            continue

        data.append({
            "title": title,
            "url": full_url
        })

    return data


# Step 1: scrape homepage
home_articles = scrape_links_from_page(BASE_URL)

df_home = (
    pd.DataFrame(home_articles)
    .drop_duplicates()
    .reset_index(drop=True)
)

print(df_home.head())
print("Total links found:", len(df_home))

# Step 2: optional scrape the discovered pages
all_articles = []

for url in df_home["url"].head(20):   # change 20 to more if needed
    page_data = scrape_links_from_page(url)
    all_articles.extend(page_data)
    time.sleep(1)

df_all = (
    pd.DataFrame(all_articles)
    .drop_duplicates()
    .reset_index(drop=True)
)

print(df_all.head())
print("Total collected:", len(df_all))

df_all.to_csv("scrapes/addis_admass_scraped_links.csv", index=False, encoding="utf-8-sig")
```
